```python
import numpy as np
import matplotlib.pyplot as plt

file_path = '/home/agirmen/Github/ColibriPipeline_SensitivityModel/sensitivity_model_output/20240729_diagnostics/Sensitivity/20240729_02.29.34.690/high_4sig_lightcurves/star0_20240729_02.29.34.690_Red.txt'

# Load the data, skipping the first 12 lines (header)
data = np.genfromtxt(file_path, skip_header=12, usecols=(1, 2))

# Separate the columns into time and flux
time = data[:, 0]
flux = data[:, 1]

print(time)
print(flux)

# Plot the data
plt.plot(time, flux)
plt.xlabel('Time (s)')
plt.ylabel('Flux')
plt.title('Lightcurve')
plt.show()
```
